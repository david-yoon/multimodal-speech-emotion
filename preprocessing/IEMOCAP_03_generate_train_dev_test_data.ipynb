## 01 generate ids
 ( delimeter = space )
- line[0] = train
- line[1] = dev
- line[2] = test

```python
import random

lines = []
with open('../data/processed/IEMOCAP/' + target_path_name + '/FC_label.txt') as f :
    lines = f.readlines()
len(lines)
```

```python
random.seed(337)

cfold01_test = []
cfold02_test = []
cfold03_test = []
cfold04_test = []
cfold05_test = []

for x in xrange( len(lines) ) :
    
    compare = random.random()
    
    if compare > 0.8 : 
        cfold01_test.append( str(x) )
    elif compare > 0.6 : 
        cfold02_test.append( str(x) )
    elif compare > 0.4 : 
        cfold03_test.append( str(x) )
    elif compare > 0.2 : 
        cfold04_test.append( str(x) )        
    else:
        cfold05_test.append( str(x) )        
```

```python
def gen_data( list_all, list_test, path ) :

    random.seed(3372)
    
    train = []
    dev = []
    test = []

    for index in xrange( len(list_all) ) :

        compare = random.random()

        if str(index) not in list_test:
            train.append( str(index) )
        else:
            if compare > 0.70 : 
                dev.append( str(index) )
            else:
                test.append( str(index) )

    with open(path, 'w') as f :
        f.write( ' '.join(train) + '\n')
        f.write( ' '.join(dev) + '\n')
        f.write( ' '.join(test) + '\n')

    print len(train)
    print len(dev)
    print len(test)
```

```python
from file_util import *
create_folder('../data/processed/IEMOCAP/'+target_path_name+'/audio_woZ_set01')
create_folder('../data/processed/IEMOCAP/'+target_path_name+'/audio_woZ_set02')
create_folder('../data/processed/IEMOCAP/'+target_path_name+'/audio_woZ_set03')
create_folder('../data/processed/IEMOCAP/'+target_path_name+'/audio_woZ_set04')
create_folder('../data/processed/IEMOCAP/'+target_path_name+'/audio_woZ_set05')
```

```python
path = '../data/processed/IEMOCAP/' + target_path_name + '/audio_woZ_set01/audio_woZ_set01.txt'
gen_data( lines, cfold01_test, path )

path = '../data/processed/IEMOCAP/' + target_path_name + '/audio_woZ_set02/audio_woZ_set02.txt'
gen_data( lines, cfold02_test, path )

path = '../data/processed/IEMOCAP/' + target_path_name + '/audio_woZ_set03/audio_woZ_set03.txt'
gen_data( lines, cfold03_test, path )

path = '../data/processed/IEMOCAP/' + target_path_name + '/audio_woZ_set04/audio_woZ_set04.txt'
gen_data( lines, cfold04_test, path )

path = '../data/processed/IEMOCAP/' + target_path_name + '/audio_woZ_set05/audio_woZ_set05.txt'
gen_data( lines, cfold05_test, path )
```
